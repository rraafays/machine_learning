```python
import pandas as pd
iris_dataset = pd.read_table('iris.txt',sep=',')
iris_dataset.head()
```

```python
import seaborn
seaborn.jointplot(data=iris_dataset, x='sepal_length', y='sepal_width',hue='species')
```

```python
seaborn.boxplot(x='species',y='petal_length',data=iris_dataset)
```

```python
pd.plotting.parallel_coordinates(iris_dataset, 'species')
```

```python
seaborn.pairplot(data=iris_dataset, hue='species')
```

```python
import numpy as np
corr = iris_dataset[['sepal_length','sepal_width','petal_length','sepal_width']].corr()
cmap = seaborn.diverging_palette(230, 20, as_cmap=True)
mask = np.triu(np.ones_like(corr, dtype=bool))# Generate a mask for the upper triangle
seaborn.heatmap(corr,cmap=cmap,mask = mask)
```
